###### FAISS

facebook Ai Similarity search(Faiss) is a library for efficient similairty search and clustering of dense vectors. It contains algorithm that search in sets of any size, up to ones that psossibly do not fit in RAM. It also contains supoorting code for evaluation and parameter tunning

```python
from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import CharacterTextSplitter
```

```python
loader = TextLoader("speech.txt")
documents= loader.load()
text_splitter= CharacterTextSplitter(chunk_size=500, chunk_overlap=30)
docs = text_splitter.split_documents(documents)
```

```python
docs
```

```python
embeddings= OllamaEmbeddings()
db= FAISS.from_documents(docs,embeddings)
```

```python
db
```

```python
##### qureing 
qurey = "we scale our applications, efficient"

docs= db.similarity_search(qurey)
docs[0].page_content
```

#### retriver
We can also convert the vectorstore into a Retriever class. Thsi allows us to easilty use it in other LangChain methods, which largely work with retrievers

```python
retriever= db.as_retriever()
docs = retriever.invoke(qurey)
docs[0].page_content
```

##### similarity search with score

```python
docs_and_score = db.similarity_search_with_score(qurey)
docs_and_score
```

```python
embeddings_vector = embeddings.embed_query(qurey)
embeddings_vector
```

```python
docs_score = db.similarity_search_by_vector(embeddings_vector)
docs_score
```

```python
### saving and loading 
db.save_local("faiss_index")
```

```python
new_df = FAISS.load_local("faiss_index",embeddings, allow_dangerous_deserialization=True)
```
